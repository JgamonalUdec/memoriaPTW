```python
import open3d as o3d
import os

# Define the path to your .ply file
# Replace "path/to/your/Memoria/file.ply" with the actual path
# For example, '/content/drive/MyDrive/Memoria/your_file_name.ply'
file_path = './PointsToWood/data/spa04-lw_ours.ply' # Replace with your file path
#file_path = './data/output/cameroon_lw_ours.ply'  

# Check if the file exists
if not os.path.exists(file_path):
    print(f"Error: File not found at {file_path}")
else:
    # Read the .ply file
    try:
        pcd = o3d.io.read_point_cloud(file_path)

        # Print basic information about the loaded point cloud
        print(f"Successfully read file: {file_path}")
        print(f"Number of points: {len(pcd.points)}")
        #o3d.visualization.draw_geometries([pcd])
        o3d.visualization.draw_plotly([pcd])

        if pcd.has_normals():
            print("Point cloud has normals.")
        if pcd.has_colors():
            print("Point cloud has colors.")

    except Exception as e:
        print(f"An error occurred while reading the file: {e}")
```
